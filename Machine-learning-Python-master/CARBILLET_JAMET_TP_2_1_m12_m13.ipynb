```python
# ESTHER JAMET & ALEXIS CARBILLET
# Assignment 12

# import des bibliothÃ¨ques
from sklearn.datasets import load_iris
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt

# import des donnÃ©es
iris = load_iris()
X = iris.data
y = iris.target

# FastICA
ica = FastICA(n_components=2, whiten=True)
ica.fit(X)

X_ica = ica.transform(X)

# Visualisation
target_ids = range(len(iris.target_names))
plt.figure()
for i, c, label in zip(target_ids, 'rgbcmykw', iris.target_names):
    plt.scatter(X_ica[y == i, 0], X_ica[y == i, 1], c=c, label=label)
plt.legend()
```

On obtient une sortie de ICA trÃ¨s proche de celle obtenue en TP par PCA.

```python
# ESTHER JAMET & ALEXIS CARBILLET
# Assignment 12

# import des bibliothÃ¨ques
from sklearn.datasets import load_iris
from sklearn.cross_decomposition import CCA
import matplotlib.pyplot as plt

# import des donnÃ©es
iris = load_iris()
X = iris.data
y = iris.target

# CCA
cca = CCA(n_components=2, scale=True)
cca.fit(X, y)

X_cca = cca.transform(X)

# Visualisation
target_ids = range(len(iris.target_names))
plt.figure()
for i, c, label in zip(target_ids, 'rgbcmykw', iris.target_names):
    plt.scatter(X_cca[y == i, 0], X_cca[y == i, 1], c=c, label=label)
plt.legend()
```

Pour CCA, la rÃ©duction de dimensionnalitÃ© a un autre aspect car les donnÃ©es sont projetÃ©es sur 1 dimension.
Cependant, on a toujours 3 clusters, qui ressemblent Ã  ceux obtenus par PCA et ICA.

```python
# ESTHER JAMET & ALEXIS CARBILLET
# Assignment 13

# import des bibliothÃ¨ques
from sklearn.datasets import make_swiss_roll
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# import des donnÃ©es
X, y = make_swiss_roll(n_samples=4000, noise=0.2)

# PCA
X_pca = PCA(n_components=2).fit_transform(X)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
```

```python
# ESTHER JAMET & ALEXIS CARBILLET
# Assignment 13

# import des bibliothÃ¨ques
from sklearn.datasets import make_swiss_roll
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt

# import des donnÃ©es
X, y = make_swiss_roll(n_samples=4000, noise=0.2)

# FastICA
X_ica = FastICA(n_components=2).fit_transform(X)
plt.scatter(X_ica[:, 0], X_ica[:, 1], c=y)
```

```python
# ESTHER JAMET & ALEXIS CARBILLET
# Assignment 13

# import des bibliothÃ¨ques
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding, Isomap
import matplotlib.pyplot as plt

# import des donnÃ©es
X, y = make_swiss_roll(n_samples=4000, noise=0.2)

# Manifold learning technique
lle = LocallyLinearEmbedding(n_neighbors=15, n_components=2, method='modified')
X_lle = lle.fit_transform(X)
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=y);
```
